```python
def get_aggs(df, col):
    drought_sum = df[col].sum()
    drought_mean = df[col].mean()
    drought_max = df[col].max()
    drought_min = df[col].min()
    drought_var = df[col].var()
    drought_std = df[col].std()
    print(f" sum : {drought_sum}\n mean: {drought_mean}\n max: {drought_max}\n min : {drought_min}\n var: {drought_var}\n std: {drought_std}")
```

```python
get_aggs(totals_for_the_years, 'tonnes per region')
```

# Africa

```python
africa_list = [  
            'Algeria', 'Angola', 'Benin', 'Botswana', 'Burkina Faso', 'Burundi', 'Cabo Verde', 'Cameroon', 
            'Central African Republic (CAR)', 'Chad', 'Comoros', 'Congo, Democratic Republic of the', 
            'Congo, Republic of the', "Cote d'Ivoire", 'Djibouti', 'Egypt', 'Equatorial Guinea', 'Eritrea',
            'Eswatini', 'Ethiopia', 'Gabon', 'Gambia', 'Ghana', 'Guinea', 'Guinea-Bissau', 'Kenya', 'Lesotho',
            'Liberia', 'Libya', 'Madagascar', 'Malawi', 'Mali', 'Mauritania', 'Mauritius', 'Morocco', 'Mozambique',
            'Namibia', 'Niger', 'Nigeria', 'Rwanda', 'Sao Tome and Principe', 'Senegal', 'Seychelles', 'Sierra Leone',
            'Somalia', 'South Africa', 'South Sudan', 'Sudan', 'Tanzania', 'Togo', 'Tunisia', 'Uganda', 'Zambia', 'Zimbabwe'
         ]
```

```python
def grouping_countries(df, my_list, col1, col2,):
     return df[df[col1].isin(my_list)].groupby(col2).sum().reset_index()
```

```python
africa_df = grouping_countries(co2_by_country_per_capita, africa_list, 'Entity', 'Year')
africa_df
```

```python
get_aggs(africa_df, 'tonnes per region')
```

```python
africa_df.info()
```

```python
africa_df.plot(x ='Year', y='tonnes per region', kind = 'line')
plt.show()
```

```python
africa_df.hist()
```
